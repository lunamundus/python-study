# 함수(function)

## 함수란?

함수는 입력값을 가지고 어떤 일을 수행한 다음, 그 결과물을 내어놓는 것을 의미함.  

수학에서 <code>y = 2x + 3</code> 이라는 함수가 있다면, y가 결과값(결과물), x가 입력값이 되는 것.  

프로그래밍에서 종종 같은 내용을 반복해서 코딩하는 경우가 있는데, 이런 경우 반복적으로 사용되는 부분을 묶어서 하나의 함수로 표현할 수 있다. 즉, 중간 과정은 똑같고 어떤 입력값을 주느냐에 따라 서로 다른 결과값을 돌려주는 동일한 코드가 바로 함수라는 것이다.  

함수를 사용해서 코드를 작성하면 프로그램의 흐름을 파악하기 쉽고, 오류가 나는 부분 역시 찾기 쉽다.

## 파이썬 함수 구조

```python
# def 함수명(매개변수):
#     수행할 문장 1
#     수행할 문장 2
#     ...
```

<code>def</code>는 함수를 만들 때 사용하는 예약어  

함수명은 만드는 사람이 임의로 지정 가능  

매개변수는 이 함수에 입력으로 전달되는 값을 받는 변수  

함수를 정의한 후, 함수 안에서 실행될 코드(문장, statement)를 작성

```python
# ex) 덧셈 함수

def add(a, b):
    return a + b
```

함수의 이름 : add  

매개변수 : a, b  

리턴값(or 출력값, 반환값) : a + b  

=> add 함수는 a와 b를 입력받고, a+b를 리턴하는 함수!

```python
# ex) add 함수 사용

# add 함수 정의
def add(a, b):
    return a + b

# add 함수 사용
print(add(1, 6))
```

```python
# ex) add 함수 사용

# add 함수 정의
def add(a, b):
    return a + b

# add 함수 사용
c = add(1, 6)

print(c)
```

## 매개변수 & 인수

매개변수는 함수에 입력으로 전달된 값으로 받는 변수  

인수는 함수를 호출할 때 전달하는 입력값

```python
def add(a, b): # a, b는 매개변수
    return a + b

print(add(1, 6)) # 1, 6은 인자
```

## 함수의 형태

### 일반적인 함수

입력값(매개변수)이 있고, 리턴값(반환값)이 있는 함수  

```python
# def 함수이름(매개변수):
#     수행할 문장 1
#     수행할 문장 2
#     ...
#     return 리턴값
```

```python
def add(a, b):
    result = a + b
    return result
```

```python
def add(a, b):
    result = a + b
    return result

c = add(3, 5) # 리턴값이 변수 c에 대입됨
print(c)
```

### 입력값(매개변수)이 없는 함수

```python
# def 함수이름():
#     수행할 문장 1
#     수행할 문장 2
#     ...
#     return 리턴값
```

```python
def print_hi():
    return 'hi'
```

```python
def print_hi():
    return 'hi'

c = print_hi() # 변수 c에 print_hi() 함수의 리턴값인 'hi' 대입 / 매개변수가 없는 함수는 인자로 넘겨주는 값이 없음
print(c)
```

### 리턴값이 없는 함수

리턴값이 없는 함수는 호출하면 리턴하는 값 없이 함수 내에 있는 문장(Statement)만 수행

```python
def non_return(a, b):
    print(f"{a+b} << 이건 리턴값이 아닙니다.")

non_return(3, 4)
```

```python
def non_return(a, b):
    print(f"{a+b} << 이건 리턴값이 아닙니다.")

a = non_return(3, 4) # 리턴값이 없기 때문에 a에 대입되는 값이 없다.
print(a)
```

### 입력값 & 리턴값 모두 없는 함수

```python
def non_input_return():
    print("입력값 & 리턴값 모두 없는 함수입니다.")

non_input_return()
```

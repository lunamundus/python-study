# while문의 기본 구조

반복해서 어떤 코드를 실행해야 하는 경우, while문을 사용한다.  
while문은 반복문이라고도 한다.  

```python
# while문의 기본 구조

# while 조건문:
#     수행할 문장 1
#     수행할 문장 2
#     ...
```

```python
a = 0


while a < 10:
    print(f"현재 a의 값은 {a}입니다.")
    a += 1

print("반복문이 종료되었습니다.")
print(f"반복문이 종료된 a의 값은 {a}입니다.")
```

## while문 만들기

```python
number = 10
input_num = 0

while input_num != number:
    input_num = int(input("숫자를 맞춰주세요 : "))

print("정답입니다.")
```

```python
# up-down 만들기
number = 10
input_num = 0

while input_num != number:
    input_num = int(input("숫자를 맞춰주세요 : "))
    
    if input_num > number:
        print("Down")
    elif input_num < number:
        print("Up")

print("정답입니다.")
```

## while문 강제로 빠져나가기

앞서 만든 up-down에서 중간에 중단하고 싶다면?  
특정 문자열을 입력받는다면 반복문을 중단하도록 제작할 수 있다.

```python
# up-down 만들기
number = 10
input_num = 0

while int(input_num) != number:
    input_num = input("숫자를 맞춰주세요 : ")

    if input_num == "q":
        print("프로그램을 종료합니다.")
        break
    
    if int(input_num) > number:
        print("Down")
    elif int(input_num) < number:
        print("Up")
    else:
        print("정답입니다.")
```

## while문의 맨 처음으로 돌아가기

특정 조건일 때 while 중간에 중단하는 것이 가능했다.  
그렇다면, 특정 조건일 때 뒤 statement를 실행하지 않고 다시 맨 처음으로 돌아가는 방법은?

```python
# 10 이하의 짝수만 출력
num = 0

while num <= 10:
    num += 1
    if num % 2 == 0:
        print(num)
```

```python
num = 0

while num <= 10:
    num += 1
    if num % 2 == 1:
        continue # break는 while문을 멈췄다면, continue는 다시 맨 앞으로 돌아간다.
    print(num)
    
```

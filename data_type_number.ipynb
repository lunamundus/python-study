# 숫자형 데이터 타입이란?
- 숫자형 데이터 타입(Numeric Data Type)은 숫자 형태로 이루어진 자료형
- 정수, 실수, 2진수, 8진수, 16진수 등 숫자로 이루어진 모든 것을 포함

# 숫자형 데이터 타입 종류

## 정수형(Integer) 데이터 타입
- 숫자형 데이터 타입 중에서 정수를 의미하는 데이터 타입
- 10, -10, 0, 1, 5 등이 해당

```python
a = 10
b = -10
c = 0
d = 1
```

## 실수형(Float, Floating-Point) 데이터 타입
- 숫자형 데이터 타입 중에서 소수점이 포함된 숫자(소수)를 의미하는 데이터 타입
- 3.14, -1,23, 1.00 등이 해당
- 컴퓨터식 지수 표현으로 선언하는 것도 가능

```python
a = 3.14
b = -1.00

# 컴퓨터식 지수표현
c = 3.12e3     # 3.12e3 = 3120.0
d = 4.25e-2    # 4.25e-2 = 0.0425
```

## 8진수(Octal) 데이터 타입
- 숫자형 데이터 타입 중에서 8진수(0~7)로 표현된 숫자를 의미하는 데이터 타입
- 0o 혹은 0O(숫자 0과 알파벳 소문자 o 혹은 대문자 O)로 시작
- 0o34, 0o72 등이 해당

```python
a = 0o34     # 0o34 = 28
b = 0o72     # 0o72 = 58
```

## 16진수(Hexadecimal) 데이터 타입
- 숫자형 데이터 타입 중에서 16진수(0~9, A~F)로 표현된 숫자를 의미하는 데이터 타입
- 문자는 각각 A = 10, B = 11, … , F = 15를 의미
- 0x 혹은 0X로 시작
- 0xFF, 0x7FC 등이 해당

```python
a = 0xFF     # 0xFF = 255
b - 0x7FC    # 0x7FC = 2044
```

# 숫자형 연산자(Arithmetic Operator)
- 숫자형 데이터 타입의 연산을 하는 연산자
- 사칙연산, 몫 반환 나누기 연산자, 나머지 반환 나누기 연산자, 거듭제곱 등이 포함

## 사칙 연산
- 가장 기본이 되는 연산자
- 더하기(+), 뺴기(-), 곱하기(*), 나누기(/) 연산자가 해당
- 나머지 연산자의 경우 정수형 데이터 타입을 가지는 변수더라도 실수형 데이터 타입으로 결과 값을 반환
- 이외의 연산자는 변수의 데이터 타입에 따라 결과 값을 반환

```python
a = 4
b = 5

print(a + b)      # 9
print(a - b)      # -1
print(a * b)      # 20
print(a / b)      # 0.8
```

```python
a = 3.14
b = 2.14

print(a + b)
print(a - b)
print(a * b)
print(a / b)
```

## 몫 반환 나누기 연산자
- 나누기 연산자 중 하나로, 몫만 반환하는 연산자
- 연산자 기호는 //로 사용
- 변수 데이터 타입에 따라 결과 값의 데이터 타입이 정해짐

```python
a = 10
b = 3

print(a // b)     # 3
```

```python
a = 10.0
b = 3.0

print(a // b)     # 3.0
```

## 나머지 반환 나누기 연산자
- 나누기 연산자 중 하나로, 나머지만 반환하는 연산자
- 연산자 기호는 %로 사용

```python
a = 10
b = 3

print(a % b)     # 1
```

```python
a = 10.0
b = 3.0

print(a % b)     # 1.0
```

## 거듭제곱 연산자
- a의 b승(a^b)을 계산하는 연산자
- 연속된 곱하기 혹은 반복문을 이용하여 계산도 가능하지만, ** 기호를 이용하여 거듭제곱을 계산할 수 있음
- 변수 데이터 타입에 따라 결과 값의 데이터 타입이 정해짐

```python
a = 2
b = 3

print(a ** b)     # 8
```

```python
a = 2.0
b = 3.0

print(a ** b)     # 8.0
```

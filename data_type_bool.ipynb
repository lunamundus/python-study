# Bool  
불(bool) 자료형이란, 참(True)과 거짓(False)을 타나내는 자료형  
불 자료형은 True와 False 2가지 값만을 가질 수 있음

```python
a = True
b = False

print(type(a))
print(type(b))
```

## 자료형의 참과 거짓  
자료형(Data Type)에 따라 참과 거짓이 존재

### 문자열(String)

```python
dt_string = "python"
print(bool(dt_string)) # True

dt_string = ""
print(bool(dt_string)) # False
```

### 리스트(List)

```python
dt_list = [1, 2, 3]
print(bool(dt_list))

dt_list = []
print(bool(dt_list))
```

### 튜플(Tuple)

```python
dt_tuple = (1, 2, 3)
print(bool(dt_tuple))

dt_tuple = ()
print(bool(dt_tuple))
```

### 딕셔너리(Dictionary)

```python
dt_dict = {a:1, b:2}
print(bool(dt_dict))

dt_dict = {}
print(bool(dt_dict))
```

### 숫자(Number)

```python
dt_num = 1
print(bool(dt_num))

dt_num = 0
print(bool(dt_num))
```

### 기타

```python
a = None

print(bool(a))
```

# Bool Operator
불 연산자는 논리(True와 False)에 대한 연산을 실행  
Python에서는 and, or, not 연산자가 구현되어 있으며, 이를 활용해서 xor, nand, nor를 구현하는 것이 가능  

## and (논리곱 연산자)
모두 참(True)일 때만 True를 반환하고, 나머지는 False를 반환한다.  
True and True = True  
True and False = False  
False and True = False  
False adn False = False  

```python
a = True
b = True

print(a and b)
```

```python
a = True
b = False

print(a and b)
```

```python
a = False
b = True

print(a and b)
```

```python
a = False
b = False

print(a and b)
```

## or (논리합 연산자)
둘 중에 하나라도 참(True)이면 True를 반환하고, 나머지는 False를 반환한다.  
True or True = True  
True or False = True  
False or True = True  
False or False = False  

```python
a = True
b = True

print(a or b)
```

```python
a = True
b = False

print(a or b)
```

```python
a = False
b = True

print(a or b)
```

```python
a = False
b = False

print(a or b)
```

## not (논리 부정 연산자)
참(True)은 거짓(False)을, 거짓(False)은 참(True)을 반환한다.  
not True = False  
not False = True

```python
a = True

print(not a)
```

```python
a = False

print(not a)
```

## xor (논리 배타적 or 연산자)

```python
def XOR(a, b):
    if a == b:
        return False
    else:
        return True

print("True xor True")
print(XOR(True, True))
print()

print("True xor False")
print(XOR(True, False))
print()

print("False xor True")
print(XOR(False, True))
print()

print("False xor False")
print(XOR(False, False))
```

## nand 연산자

```python
def NAND(a, b):
    if a == True and b == True:
        return False
    else:
        return True
    
print("True nand True")
print(NAND(True, True))
print()

print("True nand False")
print(NAND(True, False))
print()

print("False nand True")
print(NAND(False, True))
print()

print("False nand False")
print(NAND(False, False))
```

## nor 연산자

```python
def NOR(a, b):
    if a == False and b == False:
        return True
    else:
        return False
    
print("True nor True")
print(NOR(True, True))
print()

print("True nor False")
print(NOR(True, False))
print()

print("False nor True")
print(NOR(False, True))
print()

print("False nor False")
print(NOR(False, False))
```

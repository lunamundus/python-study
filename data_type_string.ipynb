# 문자형 데이터 타입이란?
- 문자형 데이터 타입(String Data Type)은 문자, 단어 등으로 구성된 문자들의 집합을 의미
- 따옴표(') 혹은 큰 따옴표(")로 둘러싸여 있으면 모두 문자열 데이터 타입으로 취급

아래 예시는 모두 문자형 데이터 타입으로 취급  
특히, 숫자의 경우 따옴표로 둘러싸여 있는지에 따라 숫자형 데이터 타입인지 문자형 데이터 타입인지 구분됨

```python
"Lorem ipsum dolor sit amet, consectetur adipiscing elit."

"a"

"123" # 따옴표 안에 들어있는 숫자는 숫자형이 아니라 문자열이다.
```

# 문자열 선언 방법
- 문자열은 따옴표 안에 있으면 모두 문자열 데이터 타입으로 선언됨
- 단, 시작과 끝 따옴표는 동일해야 함

## 작은 따옴표 이용하기
- 작은 따옴표(')를 이용하여 문자열을 선언

```python
a = 'Lorem ipsum dolor sit amet, consectetur adipiscing elit.'
```

## 큰 따옴표 이용하기
- 큰 따옴표(")를 이용하여 문자열을 선언

```python
a = "Lorem ipsum dolor sit amet, consectetur adipiscing elit."
```

## 작은 따옴표 3개 이용하기
- 작은 따옴표 3개(''')를 이용하여 문자열을 선언
- 작은 따옴표 3개를 선언할 문자열 앞 뒤로 3개씩 입력

```python
a = '''Lorem ipsum dolor sit amet, consectetur adipiscing elit.'''
```

- 작은 따옴표 3개를 이용하여 문자열을 선언하면 개행이 가능

```python
a = '''Lorem ipsum dolor sit amet, 
consectetur adipiscing elit.'''
```

- 개행이 가능하다는 특징으로, 주석으로 사용하기도 함

```python
'''
Lorem ipsum dolor sit amet, 
consectetur adipiscing elit.
'''
```

## 큰 따옴표 3개 이용하기
- 큰 따옴표 3개(""")를 이용하여 문자열을 선언
- 큰 따옴표 3개를 선언할 문자열 앞 뒤로 3개씩 입력

```python
a = """Lorem ipsum dolor sit amet, consectetur adipiscing elit."""
```

- 큰 따옴표 3개를 이용하여 문자열을 선언하면 개행이 가능

```python
a = """Lorem ipsum dolor sit amet, 
consectetur adipiscing elit."""
```

- 개행이 가능하다는 특징으로, 주석으로 사용하기도 함

```python
"""
Lorem ipsum dolor sit amet, 
consectetur adipiscing elit.
"""
```

# 문자열 안에 따옴표 포함하기
- 기본적으로 문자열은 시작과 끝에 따옴표가 위치하고 있어야 하기 때문에 따옴표를 문자열에 포함하는 것은 불가능
- 작은 따옴표로 문자열을 선언하는 경우, 큰 따옴표를 문자열에 포함할 수 있음
- 반대로 큰 따옴표로 문자열을 선언하는 경우, 작은 따옴표를 문자열에 포함할 수 있음

```python
a = 'Lorem ipsum dolor "si amet, consectetur" adipiscing elit.'

b = "Lorem ipsum dolor si's amet, consectetur adipiscing elit."
```

- 백슬래쉬(\\)를 이용하여 문자열에 포함할 수 있음(백슬래쉬 뒤에 있는 따옴표는 문자열로 취급)

```python
a = 'Lorem ipsum dolor si\'s amet, consectetur adipiscing elit.'
```

# 문자형 연산자
- 문자형 연산자(String Operator)는 덧셈(+)과 곱셈(*)이 가능

## 문자열 연결
- 더하기(+) 기호를 이용하여 두 문자열을 연결
- 문자열 덧셈이라고 부르기도 함

```python
a = 'Lorem ipsum dolor sit amet' + ' consectetur adipiscing elit.'
# a = 'Lorem ipsum dolor sit amet consectetur adipiscing elit.'
```


- 공백(띄어쓰기)도 하나의 문자열로 인식
- 공백이 필요한 연결이라면, 공백도 포함하여 연결해야 함

## 문자열 반복
- 곱하기(*) 기호를 이용하여 특정 문자열을 반복할 수 있음
- 문자열 곱셈이라고 부르기도 함

```python
a = 'hello ' * 5
# = 'hello hello hello hello hello '
```

## String Indexing

```python
a = "Hello World"
print(a[-3])
```

## String Slicing

```python
a = "Hello World"
a1 = a[0:5]
a2 = a[6:11]

print(a1)
print(a2)
```

```python
a = "Hello World"

a1 = a[0:5]
a2 = a[0:6]

print(a1)
print(a2)
```

```python
a = "Hello World"

a1 = a[0:5]
a2 = a[0:6]

print(a1)
print(a2)

if a1 == a2:
    print("a1과 a2는 같습니다.")
else:
    print("a1과 a2는 같지 않습니다.")
```

```python
a = "Hello World"

a1 = a[:5] # index 0~4까지 출력
a2 = a[6:] # index 6부터 끝까지 출력
a3 = a[:] # 문자열 전체 복사

print(a1)
print(a2)
print(a3)
```

```python
date = "20230420"

year = date[0:4]
month = date[4:6]
day = date[6:8]

print(f"오늘은 {year}년 {month}월 {day}일 입니다.")
```

## String Formatting

```python
year = 2023

day_str = "올해는 %d년 입니다." %year

print(day_str)
```

```python
year = 2023

day_str = "올해는 %s년 입니다." %year

print(day_str)
```

```python
year = 2023
month = 4
day = 22

day_str = "오늘은 %d년 %d월 %d일 입니다." %(year, month, day)

print(day_str)
```

```python
a = 3.14159265358979

pi_ = "파이 값은 %.2f 이다." %a

print(pi_)
```

```python
a = 3.14159265358979

pi_ = "파이 값은 %.4f 이다." %a

print(pi_)
```

```python
day_str = "오늘은 %4d년 %2d월 %2d일 입니다." %(2023, 1, 5)
print(day_str)

day_str = "오늘은 %4d년 %2d월 %2d일 입니다." %(2023, 10, 15)
print(day_str)
```

```python
day_str = "오늘은 %4d년 %-2d월 %-2d일 입니다." %(2023, 1, 5)
print(day_str)

day_str = "오늘은 %4d년 %-2d월 %-2d일 입니다." %(2023, 10, 15)
print(day_str)
```

```python
year = 2023

day_str = "올해는 {0}년 입니다.".format(year)

print(day_str)
```

```python
year = 2023
month = 4
day = 22

day_str = "오늘은 {0}년 {1}월 {2}일 입니다.".format(year, month, day)

print(day_str)
```

```python
year = 2023
month = 4
day = 22

day_str = "오늘은 {0}년 {0}월 {0}일 입니다.".format(year, month, day)

print(day_str)
```

```python
day_str = "오늘은 {year}년 {month}월 {day}일 입니다.".format(year=2023, month=4, day=22)

print(day_str)
```

```python
today_year = 2023
today_month = 4
today_day = 22

day_str = "오늘은 {year}년 {month}월 {day}일 입니다.".format(year=today_year, month=today_month, 
                                                            day=today_day)

print(day_str)
```

```python
print("{0:>10}".format("hi")) # 10자리, 오른쪽 정렬
print("{0:<10}".format("hi")) # 10자리, 왼쪽 정렬
print("{0:^10}".format("hi")) # 10자리, 가운데 정렬
```

```python
year = 2023
month = 4
day = 22

print(f"오늘은 {year}년 {month}월 {day}일 입니다.")
```

```python
year = 2023
month = 1
day = 22

day_str = f"오늘은 {year}년 {month}월 {day}일 입니다."

print(day_str)
```

```python
print(f'{"hi":>10}') # 10자리, 오른쪽 정렬
print(f'{"hi":<10}') # 10자리, 왼쪽 정렬
print(f'{"hi":^10}') # 10자리, 가운데 정렬
```
